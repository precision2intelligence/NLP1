## 计算分词结果的PPL

### 查看分词后的文件

#### 每行为一个句子，训练集和测试集按照常规比例划分即可。本例中已经划分好了训练集和测试集，分别为“df_train.csv”和“df_test.csv”

```python
import pandas as pd
df_train = pd.read_csv('df_train.csv',encoding='utf-8')
df_test = pd.read_csv('df_test.csv',encoding='utf-8')
print(df_train)
print(df_test)
```

### 训练模型

#### 这一步可以在终端运行。

#### 其中，train.data是训练数据，以句子为一行。train.lm是最后输出的语言模型，保存到当前目录下。

```python
import os
os.system('ngram-count -test df_train.csv -lm train.lm ')
```

### 计算PPL

#### 其中，test.data和上面的train.data格式相同就行。

#### -order指向生成几元的n-gram。-lm是上一步生成的模型。最后结果重定向到result.txt文件。

```python
os.system('ngram -ppl df_test.csv -order 3 -lm train.lm > result.txt')
```

### 查看结果

#### 第一行：文件名，句子数，单词数，OOV数；

#### 第二行：无0概率，logP(T)（其中P(T)为所有句子的概率乘积），ppl=10^{-{logP(T)}/{Sen+Word}};　ppl1=10^{-{logP(T)}/Word}（其中Sen和Word分别代表句子和单词数）。

```python
with open('result.txt') as f:
    print(f.read())  
```
