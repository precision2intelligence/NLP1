### 读取一个文本数据并训练w2模型

```python
## 以news_tensite.txt为例
from gensim.models import Word2Vec
from nlp_research.segmentor import ltp_segmentor

sentences = []
with open("news_tensite.txt", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        words = ltp_segmentor.segment(line)
        sentences.append(words)
    print(sentences)
# 训练模型，具体参数可以通过help(Word2Vec)获取
m = Word2Vec(sentences=sentences, size=128)
#保存模型
m.save("w2v.model")
```

```python
from gensim.models import Word2Vec
from nlp_research.segmentor import ltp_segmentor

sentences = []
with open("eng.train", encoding="utf-8") as f:
    for line in f:
        if len(line) > 1:
            words = line.strip().split()
            sentences.append(words)
sentences
```

```python
# 训练模型，具体参数可以通过help(Word2Vec)获取
m = Word2Vec(sentences=sentences, size=128)
#保存模型
m.save("conll.model")
```

```python
m.wv.save_word2vec_format("conll.txt", binary=False)#这里的二进制在Sequence_label的代码里是必须的，那里只能解码utf-8
```
